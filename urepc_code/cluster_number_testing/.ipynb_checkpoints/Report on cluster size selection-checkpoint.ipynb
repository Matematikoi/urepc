# Finding the amount of clusters

The initial setup was to look for no more than 4 cluster but the results showed that there was more than 4.

## Setup

We have 2 databases 1.5 and 2.0 and we want to focus only on particular rows (all the rows that end with "\_1","\_2" or "\_3"). This gives us $2\cdot 3=6$ datasets. But for each dataset we can make or not a "log transformation" consisting in $f(x)=\log(x+1)$. This gives us a total of $12$ dataset to analyze.

## Try with $k = 5$ ( let $k$ be the max amount of clusters that ```repnormmixmodel.sel``` takes into account)

For each dataset we have 4 selection criteria, meaning that we are looking at $6\cdot4=24$ datasets

````python
import pandas as pd
import numpy as np
from IPython.display import Image
from sklearn.manifold import TSNE
import seaborn as sns
from sklearn.decomposition import PCA
````

````python
report_k_5 = pd.read_csv('./results/selection_results.csv', index_col=0)
report_k_5.iloc[:,
                np.array(~report_k_5.columns.str.startswith('X')) &
                np.array(report_k_5.columns != 'file')
               ]
````

## Datasets missing
The function ```repnormmix.sel``` will fail if it has to restart itself too many times ( maybe there is a way of making it try more or in a smarter way using ```repnormmixEM``` parameters like ```arbmean``` or ```arbvar```).


This means that none of the non log transformation "survived" the algorithm. Some other transformations/normalizations should be looked at to be sure that the results are not biased. 


Since all the results point towards $k=5$ clusters, we should look for more than $5$ clusters to be sure that the result is inside our parameter space for $k$

# Analyzing a $k=9$ cluster with ```npEM``` using ```t-SNE``` and ```PCA```

We have that $\lambda = (0.18,0.23,0.06,0.001,0.03, 0.14, 0.04 ,0.19 ,0.14)^T$

We can probably get rid of the 4 lesser values since their contribution is very small. But since we already have this selection lets analyze how it would look like

### Note :
The ```npEM``` algorithm did not converge it stop after 1000 iterations and nearly 1 day of running, results with convergence may vary. 

````python
posteriors = pd.read_csv("./results/posteriors_1_5_CPMcutoff_suffix_1_log_it_1000_npEM.csv",index_col=0)
clusters = np.argmax(posteriors.values,axis = 1)
data_points = pd.read_csv("./data/1_5_CPMcutoff_suffix_1_log.csv",index_col=0)
````

````python
%%time
X = data_points.values
X_embedded = TSNE(n_components=2).fit_transform(X)
````

````python
df = pd.DataFrame({
    'x':X_embedded[:,0],
    'y':X_embedded[:,1], 
    'cluster':clusters
})
g = sns.scatterplot(data=df ,x='x',y='y' ,hue="cluster",alpha = 0.7,palette='bright');
g.figure.set_size_inches(20,15);
g.set_title("Visualization using t-SNE",fontsize=35);
````

````python
X_pca = PCA(n_components=2).fit_transform(X)
df2 = pd.DataFrame({
    'x':X_pca[:,0],
    'y':X_pca[:,1], 
    'cluster':clusters
})
g = sns.scatterplot(data=df2 ,x='x',y='y' ,hue="cluster",alpha = 0.7,palette='bright');
g.figure.set_size_inches(20,15);
g.set_title("Visualization using PCA",fontsize=35);
````
