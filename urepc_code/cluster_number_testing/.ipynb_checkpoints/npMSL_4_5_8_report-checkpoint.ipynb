```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
sns.set_style("darkgrid")
```

# Cluster selection with cero replacement (uniform distribution)

```python
df = pd.read_csv('./results/cluster_sel.csv',index_col=0)
df *= -1
df = df.T.iloc[:-1,:]
sel = 'ICL'
df = df[df.index.astype(int) > 3]
g = sns.barplot(data = df, y = sel, x = df.index, palette = 'viridis');
g.set(ylim = (np.min(df[sel])-2,np.max(df[sel])));
g.set(title = "ICL scores on different cluster sizes");
```

# Reading the data

```python
posteriors_4 = pd.read_csv(
    './results/np_posteriors/npMSL_k_4_1_5_CPMcutoff_suffix_1_log_cero_replacement_posteriors.csv',
    index_col=0)
posteriors_5 = pd.read_csv(
    './results/np_posteriors/npMSL_k_5_1_5_CPMcutoff_suffix_1_log_cero_replacement_posteriors.csv',
    index_col=0)
posteriors_8 = pd.read_csv(
    './results/np_posteriors/npMSL_k_8_1_5_CPMcutoff_suffix_1_log_cero_replacement_posteriors.csv',
    index_col=0)

names = ['k = 4','k = 5','k = 8']
data_sets = [posteriors_4,posteriors_5,posteriors_8]
```

```python
data_points = pd.read_csv('./data/1_5_CPMcutoff_suffix_1_log_cero_replacement.csv',index_col=0)
```

# Lambda in each cluster

```python
fig, axs = plt.subplots(ncols=3,figsize=(18, 5))
cnt = 0
for name, data in zip(names, data_sets):
    data_aux = data.copy()
    data_aux['cluster'] = np.argmax(data_aux.values, axis = 1)
    g = sns.barplot(x = data_aux['cluster'].value_counts(normalize= True).index,
                    y = data_aux['cluster'].value_counts(normalize= True),
                    ax = axs[cnt], 
                    order = data_aux['cluster'].value_counts(normalize= True).index,
                    palette='viridis')
    g.set_title('Distribution of clusters {}'.format(name))
    cnt += 1
fig.suptitle("Percentage of the population in each cluster",fontsize = 25)
plt.tight_layout()
```

# Max probability in each cluster

```python
fig, axs = plt.subplots(ncols=3, nrows = 8, figsize=(18, 18))
cnt = 0
for name, data in zip(names, data_sets):
    for col, i in zip(data.columns,range(len(data.columns))):
        data_aux = data.copy()
        data_aux['cluster'] = np.argmax(data_aux.values, axis = 1)
        data_aux['max_prob'] = np.max(data_aux.values[:,:len(data.columns)], axis = 1)
        data_aux = data_aux[data_aux['cluster'] == i]
        
        g = sns.histplot(
            data = data_aux,
            x = 'max_prob',
            stat = 'probability',
            binwidth=0.05,
            ax = axs[i, cnt]
        )
        g.set_title('Cluster {}, {}'.format(i,name))
        g.set(xlim=(0, 1))
        plt.tight_layout()
    cnt += 1
fig.suptitle("Maximum probability in each cluster",fontsize = 25,y=1.0)
#fig.subplots_adjust(top=.8)
plt.tight_layout()
```

# Norm

```python
sns.histplot(x = np.linalg.norm(data_points.values, axis = 1,ord = 1)).set_title('Norm in the dataset');
```

```python
fig, axs = plt.subplots(ncols=3, nrows = 8, figsize=(18, 18))
cnt = 0
for name, data in zip(names, data_sets):
    for col, i in zip(data.columns,range(len(data.columns))):
        data_aux = data.copy()
        data_aux['cluster'] = np.argmax(data_aux.values, axis = 1)
        data_aux['norm'] = np.linalg.norm(data_points.values, axis = 1,ord = 1)
        data_aux = data_aux[data_aux['cluster'] == i]
        
        g = sns.histplot(
            data = data_aux,
            x = 'norm',
            stat = 'probability',
            ax = axs[i, cnt]
        )
        g.set_title('Cluster {}, {}'.format(i,name))
        g.set(xlim=(0, 30))
        plt.tight_layout()
    cnt += 1
fig.suptitle("Norm histogram in each cluster",fontsize = 25,y=1.0)
#fig.subplots_adjust(top=.8)
plt.tight_layout()
```

# Norm vs probability

```python
fig, axs = plt.subplots(ncols=3,figsize=(18, 5))
cnt = 0
for name, data in zip(names, data_sets):
    data_aux = data.copy()
    data_aux['max_prob'] = np.max(data_aux.values[:,:8],axis = 1)
    data_aux['norm'] = np.linalg.norm(data_points.values, axis = 1,ord = 1)
    g = sns.scatterplot(
        data = data_aux,
        x = 'norm',
        y = 'max_prob',
        ax = axs[cnt], 
        palette='viridis',
        alpha = 0.01
    )
    g.set_title('Norm vs probability {}'.format(name))
    cnt += 1
```

# t-SNE

```python
X = data_points.values
X_embedded = TSNE(n_components=2).fit_transform(X)
```

```python
df = pd.DataFrame({
    'x':X_embedded[:,0],
    'y':X_embedded[:,1]
})
fig, axs = plt.subplots(nrows=3,figsize=(18, 5))
cnt = 0
for name, data in zip(names, data_sets):
    df['cluster'] = np.argmax(data.values, axis = 1)
    g = sns.scatterplot(data=df ,x='x',y='y' ,hue="cluster",alpha = 0.7,palette='bright', ax= axs[cnt]);
    g.figure.set_size_inches(20,15);
    g.set_title("Visualization npMSL using t-SNE {}".format(name),fontsize=35);    
    cnt+=1
plt.tight_layout()
```
