```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data = pd.DataFrame({'max_k':[],'max_it':[],'AIC':[],'BIC':[],'CAIC':[],'ICL':[]})
for root, dirs, files in os.walk("../../cluster_code/results/cs_rep/"):
    for filename in files:
        if 'csv' in filename:
            temp_data = pd.read_csv(root+filename, index_col=0).Winner
            data = data.append({
                'max_k':filename.split('_')[1],
                'max_it':filename.split('_')[-1][:-4],
                'AIC':temp_data.values[0],
                'BIC':temp_data.values[1],
                'CAIC':temp_data.values[2],
                'ICL':temp_data.values[3]
            }, ignore_index=True)
```

```python
data.max_k = data.max_k.astype(int)
data.max_it = data.max_it.astype(int)
data = data[data.max_it > 100]
data.sort_values(['max_it','max_k'])
```
