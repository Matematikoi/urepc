```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('../../cluster_code/results/sem_results_2/loglik_sem_2.csv')
df = df[df['index'] != 5]
```

```python
df['BIC'] = df.loglik - 0.5*df['size']*3*np.log(16247)
df.sort_values('BIC', ascending = False)
```

```python
idx = df.groupby('size')['BIC'].transform(max) == df['BIC']
df = df[idx].sort_values('BIC', ascending = False)
df
```

```python
files = os.listdir('../../cluster_code/results/sem_results/')
files = ['../../cluster_code/results/sem_results/'+f for f in files if 'out_' in f]
```

```python
df_out = pd.DataFrame({
    'cluster_number':[],
    'loglik':[],
    'iterations':[]
})
for file in files:
    with open(file) as f:
        lines = f.readlines()
        number_clusters = int(lines[0].split()[1].split('_')[3])
        loklik = float(lines[-4].split()[-1])
        iterations = (lines[-4].split()[1])
        df_out = df_out.append({
            'cluster_number': number_clusters,
            'loglik': loklik,
            'iterations': iterations,
        },ignore_index = True)
df_out = df_out[df_out.iterations != 'of']
df_out.iterations = df_out.iterations.astype(int)
df_out
```

```python
print(len(files))
df_out = df_out[(df_out.iterations > 1001)].copy()
df_out['BIC'] = df_out.loglik - 0.5*df_out['cluster_number']*3*np.log(16247)
```

```python
df_out.sort_values('BIC', ascending = False).drop_duplicates()
```

```python
files = os.listdir('../../cluster_code/results/sem/')
def get_name (idx):
    x = [f for f in files if ('model_sem' in f and '_'+idx in f) ]
    x.sort()
    lam, mu, sigma = x
    return lam, mu, sigma
```

```python
for i in df['index']:
    lam,mu,sigma = get_name(i)
    with open('../../cluster_code/data/sem/selected_models.csv', 'a') as myfile:
        myfile.write(','.join([lam,mu,sigma])+'\n')
    
```
