```python
import pandas as pd 
from pandas_profiling import ProfileReport
import numpy as np
import seaborn as sns
```

```python
original_data = pd.read_csv('./data/fly_data_embryos.csv', index_col=0)
original_data = original_data[np.max(original_data.values, axis = 1) > 0]
original_data.head()
```

```python
data = pd.read_csv('./results/sim_data_1.csv',index_col=0)
data.columns = original_data.columns
data.head()
```

```python
data.describe()
```

```python
profile = ProfileReport(np.log(data+1), title = "Simulated data report")
```

```python
profile.to_widgets()
```

# Try with the original data

```python
profile = ProfileReport(np.log(original_data+1), title = "Original fly data report")
profile.to_widgets()
```

# Compare statistics among the databases

```python
data[np.max(data.values, axis = 1 ) > 0].describe() 
```

# Skewness and kurtosis

```python
data.kurtosis(), original_data.kurtosis()
```

```python
data.skew(),original_data.skew()
```

# Correlations

```python
method = 'pearson'
(data.corr(method = method) - original_data.corr(method = method))
```

# A little analysis 

Regarding the mean, they are almost the same. 

Regarding the standard deviation: they are almost the same

The histograms are quite similar but not completely. some tend more to zero, and some tend more to extreme values. 

Kurtosis and skewness are also similar.

Correlation is quite similar, but the original data is a little more correlated.
