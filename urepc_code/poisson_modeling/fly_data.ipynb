# Analysis of fly data

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
genes = pd.read_csv('./data/modencodefly_phenodata.txt', sep = ' ')
genes
```

```python
data = pd.read_csv('./data/modencodefly_pooledreps_count_table.txt', sep ='\t')
data.head()
```

```python
genes_chosen = genes[genes['sample.id'].isin(data.columns)].copy()
#genes_chosen.head(100)
```

```python
genes_chosen['num.tech.reps'].value_counts()
```

```python
new_names = {}
new_names_inv ={}
for idx,row in genes_chosen.iterrows():
    new_names[row['sample.id']]= row['stage']
    new_names_inv[row['stage']] = row['sample.id']
```

```python
data_embryos = data.copy()
data_embryos.rename(columns = new_names, inplace = True)
data_embryos.index = data_embryos.gene
data_embryos = data_embryos[[col for col in data_embryos.columns if 'Embryos' in col]]
```

```python
data_embryos[np.max(data_embryos.values, axis = 1) > 0]
```

```python
data_embryos.to_csv('./data/fly_data_embryos.csv')
data_embryos.rename(columns = new_names_inv, inplace = True)
data_embryos.to_csv('./data/fly_data_embryos_SRX_names.csv')
```
