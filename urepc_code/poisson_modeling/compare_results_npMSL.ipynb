# Compare the results from npMSL

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
from sklearn.metrics.cluster import adjusted_rand_score
sns.set_theme()
```

```python
files = [i for i in os.listdir('../../cluster_code/npMSLvsPoisson/results/') if 'posteriors' in i]
```

```python
labels = {}
def get_labels(idx) :
    if idx not in labels:
        labels[idx] = np.argmax(pd.read_csv('../../cluster_code/npMSLvsPoisson/results/'+files[idx],index_col=0).values, axis = 1)
    return labels[idx]
```

```python
labels_gt = pd.read_csv('./human_sim_data/labels_complete.csv').x.values
```

```python
results = pd.DataFrame({'same_bw':[],'iterations':[],'cluster_size':[], 'dataset':[], 'ngrid':[], 'ARI':[]})
for idx, value in enumerate(files):
    results = results.append(
        {
            'same_bw':value.split('_')[2],
            'iterations':value.split('_')[3],
            'cluster_size':value.split('_')[4], 
            'dataset':'.'.join(value.split('_')[5:-1]), 
            'ngrid':value.split('_')[-1][0:-4], 
            'ARI': adjusted_rand_score(labels_gt, get_labels(idx))
        }, ignore_index=True)    
```

```python
results.sort_values(by='ARI', ascending=False).head(20)
```

```python
files_npEM = [i for i in os.listdir('../../cluster_code/npMSLvsPoisson/results_npEM/') if 'posteriors' in i]
```

```python
labels_npEM = {}
def get_labels_npEM(idx) :
    if idx not in labels_npEM:
        labels_npEM[idx] = np.argmax(pd.read_csv('../../cluster_code/npMSLvsPoisson/results_npEM/'+files_npEM[idx],index_col=0).values, axis = 1)
    return labels[idx]
```

```python
results_npEM = pd.DataFrame({'same_bw':[],'iterations':[],'cluster_size':[], 'dataset':[],  'ARI':[]})
for idx, value in enumerate(files_npEM):
    print(idx, end=',')
    results_npEM = results_npEM.append(
        {
            'same_bw':value.split('_')[2],
            'iterations':value.split('_')[3],
            'cluster_size':value.split('_')[4], 
            'dataset':'.'.join(value.split('_')[5:]), 
            'ARI': adjusted_rand_score(labels_gt, get_labels_npEM(idx))
        }, ignore_index=True)    
```

```python
results_npEM.sort_values(by='ARI', ascending=False)
```

```python
files_one_col = [i for i in os.listdir('../../cluster_code/npMSLvsPoisson/results_np_one_col/') if 'posteriors' in i]
```

```python
labels_np = {}
def get_labels_np(idx) :
    if idx not in labels_np:
        labels_np[idx] = np.argmax(
            pd.read_csv('../../cluster_code/npMSLvsPoisson/results_np_one_col/'+files_one_col[idx],index_col=0).values, axis = 1)
    return labels[idx]
results_np = pd.DataFrame({'method':[],'same_bw':[],'iterations':[],'cluster_size':[], 'col':[], 'dataset':[],  'ARI':[]})
for idx, value in enumerate(files_one_col):
    print(idx, end=',')
    results_np = results_np.append(
        {
            'method':value.split('_')[2],
            'same_bw':value.split('_')[3],
            'iterations':value.split('_')[4],
            'cluster_size':value.split('_')[5],
            'col':value.split('_')[6],
            'dataset':'.'.join(value.split('_')[7:]), 
            'ARI': adjusted_rand_score(labels_gt, get_labels_np(idx))
        }, ignore_index=True) 
```
